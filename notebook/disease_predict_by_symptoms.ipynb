# Define X and Y - from training data and P for testing data

```python
P = train[["prognosis"]]
X = train.drop(["prognosis"],axis=1)
Y = test.drop(["prognosis"],axis=1)
```

# SPLITTING DATESET

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,P,test_size=0.2,random_state=42)
```

# Model  - Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dtc= DecisionTreeClassifier(random_state=42)
model_dtc = dtc.fit(xtrain,ytrain)
tr_pred_dtc = model_dtc.predict(xtrain)
ts_pred_dtc = model_dtc.predict(xtest)

print("training accuracy is:",accuracy_score(ytrain,tr_pred_dtc))
print("testing accuracy is:",accuracy_score(ytest,ts_pred_dtc))
```

# saving the traind model

```python

import pickle
filename = 'disease_predict_by_symptoms.sav'
pickle.dump(model_dtc, open(filename, 'wb'))
```
